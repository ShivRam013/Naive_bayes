```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("Naive-Bayes-Classification-Data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset,hue="diabetes")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["diabetes"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

gnb=GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
```
